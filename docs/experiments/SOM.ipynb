```python
from sklearn.datasets import make_blobs

centers = [[1, 1], [-1, -1], [1, -1], [-1, 1]]
X, _ = make_blobs(n_samples=2000000, centers=centers, cluster_std=0.6)
```

```python
import math

def memory_size(array):
    size_bytes = array.size * array.itemsize
    size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    
    return "%s %s" % (s, size_name[i])

memory_size(X)
```

```python
from minisom import MiniSom

som = MiniSom(2, 2, 2, sigma=0.3, learning_rate=0.5) # initialization of 6x6 SOM
```

```python
%time som.train(X, 100) # trains the SOM with 100 iterations
```

```python
from somoclu import Somoclu

csom = Somoclu(2, 2, compactsupport=False)
```

```python
%time csom.train(X)
```
